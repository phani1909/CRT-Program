```python
# read the two numbers as input and print the output as the number which is divisible by 3 and 5
    test cases :
        1  15 -- 15
        #include<stdio.h>
int main()
{
	int x,y;
	scanf("%d%d",&x,&y);
	while(x<=y)
	{
		if(x%3==0 && y%5==0)
		{
			printf("%d",x);
		}
		x++;
	}
	return 0;
}
        
```

```python
# read the number as input and print the output as only the sum of even digits in the given number
    test cases :
        1889 -- 16
        145 --- 4
#include<stdio.h>
int main()
{
	int x,s=0,r;
	scanf("%d",&x);
	while(x!=0)
	{
		r =x%10;
		if(r%2==0)
		{
			s=s+r;
		}
		x=x/10;
	}
	printf("%d",s);
	return 0;
}
```

```python
# read the input as a number and print the output in the following manner
     test cases:
        1889 -- 5
        666 -- 3
        4756 -- 3
        1591 -- 1
# reason for the output is each circle will cost of 1 unit and tringle cost of 2 units
#include<stdio.h>
int main()
{
	int x,r,s=0;
	scanf("%d",&x);
	while(x!=0)
	{
		r=x%10;
		if(r==8)
		{
			s += 2;
		}
		else if(r==0 || r==6 || r==9)
		{
			s += 1;
		}
		else if(r==4)
		{
			s += 2;
		}
		x/=10;
	}
	printf("%d",s);
	return 0;
}
```
